#### Ques2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

Ans.
1. Null Hypothesis (Ho):
A statement of no effect, no difference, or status quo.  
Example: "The average weight of apples is 150 grams."

2. Alternative Hypothesis (H₁ or Ha):
A statement that contradicts the null. It represents the effect or change you are testing for.  
Example: "The average weight of apples is not 150 grams."


#### Ques3. What is the significance level in hypothesis testing, and why is it important?

Ans. The significance level is the probability of rejecting the null hypothesis when it is actually true (a Type I error).It represents the maximum risk you're willing to take to falsely reject Ho.

Common α values:

0.05 (5%)

0.01 (1%)

0.10 (10%)

Importance:
It helps in interpreting the P-value (if P ≤ alpha → reject H₀).  
The P-value is the actual evidence, and α is your rule for how strong that evidence must be to reject H₀.
If the evidence (P-value) is strong enough (P ≤ α), you reject H₀.


#### Ques4. What does a P-value represent in hypothesis testing?

Ans. The P-value is the probability of obtaining results at least as extreme as the observed data, assuming that the null hypothesis is true.A small P-value (typically ≤ α) indicates that the observed data is unlikely under H₀, and thus provides evidence against H₀.  
A large P-value means the data is consistent with H₀ → do not reject.

#### Ques5. How do you interpret the P-value in hypothesis testing?


Asn. The P-value is the probability of getting a result as extreme as or more extreme than the one observed, assuming that the null hypothesis (H₀) is true.
Let’s say:  

You test if a new drug is more effective than the current one.

Null Hypothesis (H₀): New drug is not better than the current one.

You get a P-value = 0.03, and your α = 0.05.

Since 0.03 < 0.05, you reject H₀.
➡️ You conclude there is significant evidence that the new drug works better.
The smaller the P-value, the stronger the evidence against the null hypothesis.
But the P-value does not tell you the probability that H₀ is true or false.

#### Ques7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

Ans. In hypothesis testing, the choice between a one-tailed and a two-tailed test depends on the direction of the effect you are testing for.   
A one-tailed test is used when you want to check for a difference in only one specific direction. For example, if your alternative hypothesis is that the population mean (μ) is greater than a certain value (μ₀), you write it as H₁: μ > μ₀ — this is a right-tailed test. Similarly, if you're testing whether the population mean is less than μ₀, then H₁: μ < μ₀, which is a left-tailed test. In both cases, you're only concerned with results on one side of the distribution.  
On the other hand, a two-tailed test is used when you want to check for any significant difference, regardless of direction. The alternative hypothesis here is H₁: μ ≠ μ₀, meaning you're interested in whether μ is either less than or greater than μ₀. The critical region (where you reject the null hypothesis) is divided equally between both tails of the normal or t-distribution. This is more conservative because it requires stronger evidence to reject H₀ in either direction.  
One-tailed: “The new drug increases recovery rate.”  

Two-tailed: “The new drug affects recovery rate (could be better or worse).”

#### Ques14.What is a confidence interval, and how is it used to interpret statistical results?

Ans.A confidence interval (CI) is a range of values derived from sample data that is likely to contain the true population parameter (like the mean) with a specified level of confidence.
Example>>  
A 95% CI for a sample mean might be:(48.2,51.8).This means:

"We are 95% confident that the true population mean lies between 48.2 and 51.8."

It does not mean there's a 95% chance that the true mean is in the interval — it means that if we repeated the study many times, 95% of the calculated intervals would contain the true mean.






#### Ques15. What is the margin of error, and how does it affect the confidence interval?

Ans. The margin of error (ME) is the amount you're willing to allow for sampling error. It determines how wide the confidence interval is.(ME) tells us how much we expect our estimate (like a sample mean) to vary from the true population value.
Formula (simplified for mean):  
Margin of Error = Critical Value × Standard Error  
A larger margin of error results in a wider interval (more uncertainty).
When ME is big, you're saying:

"I’m not very sure where the true value is, so I need a wider range to be safe."
Example:
x =70, ME = 10 → CI = (60, 80)

A smaller margin of error results in a narrower interval (more precision).When ME is small, you're saying:

"I’m more confident in my estimate, so I can give a tighter range."

#### Ques3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.


Bernoulli Distribution Basics:
A Bernoulli distribution has only two possible outcomes:

Success (1) with probability p

Failure (0) with probability 1 - p

P(X=x)=p^x⋅(1−p)^(1−x)

where x is 0 or 1, p is the probability of success.


image.png

```python
def bernoulli_pdf(x, p):
  '''
  Calculate the PDF of a Bernoulli distribution

  Parameters:
      x (int): value, (0/1)
      p (float): Probability of success (0 to 1)

  Returns:
     float: Probability P(X=x)
  '''
  if x not in [0,1]:
    raise ValueError("x must be 0 or 1 for Bernoulli distribution.")
  if not( 0 <= p <= 1):
    raise ValueError("p must be between 0 and 1.")

  return p**x * (1-p)**(1-x)


print("P(X = 1):", bernoulli_pdf(1, 0.7))
print("P(x = 0):", bernoulli_pdf(0, 0.7))
```

#### Ques10.Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.


```python
'''If you take multiple random samples from any population (even if it's not normally distributed), the distribution of the sample means will approach a normal distribution as the sample size increases.'''
'''Generate a non-normal population (exponential distribution).

Take repeated random samples from this population.

Compute the mean of each sample.

Plot the distribution of these sample means.'''
```

```python
population = np.random.exponential(scale =2.0, size =100000)

sample_size = 50
num_samples =1000
sample_means = []

for _ in range(num_samples):
  sample = np.random.choice(population, size = sample_size)# randomly selects 50 values from the population.
  sample_mean = np.mean(sample)
  sample_means.append(sample_mean)

plt.figure(figsize=(14, 6))

# Histogram of original population (Exponential - not normal)
plt.subplot(1, 2, 1)
plt.hist(population, bins=50, color='orange', edgecolor='black')
plt.title("Original Population (Exponential)")
plt.xlabel("Value")
plt.ylabel("Frequency")


# Histogram of sample means (CLT result - approx. normal)
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=40, color='skyblue', edgecolor='black')
plt.title("Distribution of Sample Means (CLT)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

#### Ques15.Implement hypothesis testing using Z-statistics for a sample dataset.


```python
# Define parameters>>
mu = 170  # Population mean (assumed)

sample = np.random.normal(loc = 172, scale =10, size =100)

#sample size>>
n = len(sample)

#sample mean and std deviation>>
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof = 1)

z = (sample_mean - mu) / (sample_std / np.sqrt(n))
print("Z-score:", z)

# Two-tailed test: use survival function for both ends
p_value = 2*(1- stats.norm.cdf(abs(z)))
print("P-value:", p_value)

alpha = 0.05  # significance level

if p_value < alpha:
    print("Reject the null hypothesis — significant difference.")
else:
    print("Fail to reject the null — no significant difference.")

```

#### Ques16.Create a confidence interval for a dataset using Python and interpret the result.

#### Ques17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.


```python
np.random.seed(42)

data =np.random.normal(loc =50, scale= 10, size =100)

sample_mean = np.mean(data)
sample_std = np.std(data, ddof =1)
n = len(data)

standard_error = sample_std/ np.sqrt(n)

confidence_level = 0.95

z_score = stats.norm.ppf(1- (1-confidence_level) / 2)

margin_of_error = z_score * standard_error

#confidence Interval>>
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean : {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

#### Ques5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.


```python
#import numpy, matplotlib.pyplot and scipy.stats import norm

#Function definition>>
def visualize_type_errors(mu0 = 0, mu1 =1, sigma =1, alpha = 0.05, tail ='right'):

  x = np.linspace(mu0 - 4*sigma, mu1+ 4*sigma, 1000)
  y0= norm.pdf(x, mu0, sigma)
  y1 = norm.pdf(x, mu1, sigma)

  plt.figure(figsize = (10,6))
  plt.plot(x, y0, label = "H0 : Mu ={}".format(mu0), color = "blue")
  plt.plot(x, y1, label = "H1: Mu ={}".format(mu1), color = "green")

  if tail == "right":
           critical_value = norm.ppf(1 - alpha, mu0, sigma)
           plt.axvline(critical_value, color = 'red', linestyle = "--", label = "Critical Values")

           # Type I Error (α)
           x_alpha  = x[x> critical_value]
           plt.fill_between(x_alpha, norm.pdf(x_alpha , mu0, sigma), color = 'red', alpha = 0.3, label = 'Type I Error (alpha)')

           # Type II Error (β)
           x_beta = x[x< critical_value]
           plt.fill_between(x_beta, norm.pdf(x_beta, mu1 , sigma), color = 'orange', alpha = 0.3, label = 'Type II Error')

  elif tail == "left":
           critical_value = norm.ppf(alpha, mu0, sigma)
           plt.axvline(critical_value, color= "red", linestyle = "--", label = 'Critical Value')

           x_alpha = x[x < critical_value]
           plt.fill_between(x_alpha, norm.pdf(x_alpha, mu0, sigma), color='red', alpha=0.3, label='Type I Error')

           x_beta = x[x > critical_value]
           plt.fill_between(x_beta, norm.pdf(x_beta, mu1, sigma), color='orange', alpha=0.3, label='Type II Error')
  elif tail == 'two':
           crit_low = norm.ppf(alpha / 2, mu0, sigma)
           crit_high = norm.ppf(1 - alpha / 2, mu0, sigma)
           plt.axvline(crit_low, color='red', linestyle='--')
           plt.axvline(crit_high, color='red', linestyle='--', label='Critical Values')

           x_alpha_low = x[x < crit_low]
           x_alpha_high = x[x > crit_high]
           plt.fill_between(x_alpha_low, norm.pdf(x_alpha_low, mu0, sigma), color='red', alpha=0.3)
           plt.fill_between(x_alpha_high, norm.pdf(x_alpha_high, mu0, sigma), color='red', alpha=0.3, label='Type I Error (α)')

           x_beta = x[(x > crit_low) & (x < crit_high)]
           plt.fill_between(x_beta, norm.pdf(x_beta, mu1, sigma), color='orange', alpha=0.3, label='Type II Error (β)')

  else:
         raise ValueError("tail must be 'right', 'left', or 'two'")

  plt.title("Type I and Type II Error Visualization")
  plt.xlabel("Test Statistic")
  plt.ylabel("Probability Density")
  plt.legend()
  plt.grid(True)
  plt.show()
visualize_type_errors(mu0=0, mu1=1, sigma=1, alpha=0.05, tail='right')
```

#### Ques9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.


```python
#A confidence interval (CI) gives a range of values that is likely to contain the true population mean.
#95% Confidence Interval: (13.4, 15.2)
#“We are 95% confident that the true population mean lies between 13.4 and 15.2.”

from scipy import stats
import numpy as np

def confidence_interval(data, confidence = 0.95):

  # Returns:
        #(lower_bound, upper_bound): tuple of confidence interval
  data = np.array(data) #mean calculate krne k lie
  n = len(data)
  mean =np.mean(data)
  std_err = stats.sem(data)

  #t-scritical value for two-tailed t-distribution>
  #This t_crit is later used to calculate the margin of error in a confidence interval:
  t_crit = stats.t.ppf((1+ confidence)/2 , df = n-1)
  #ppf>> gives you the t-value such that area to the left = p,>> stats.t.ppf(p, df)


  margin_of_error = t_crit * std_err
  lower = mean - margin_of_error
  upper = mean + margin_of_error

  return (lower, upper)


sample = [12, 14, 15, 13, 16, 14, 15, 13, 14, 16]
ci = confidence_interval(sample)
print(f"95% Confidence Interval: {ci}")
```

 Significance of Confidence Interval:
 *  Estimates the Population Mean:
 Instead of guessing a single value (called a point estimate), a CI gives a range of possible values where the true mean likely falls.

 * Shows the Precision of the Estimate:
narrow interval → More precise estimate (less uncertainty)
A wide interval → More uncertainty in the estimate

*  Better than P-values Alone:
Unlike p-values (which only tell you if something is "significant" or not), confidence intervals tell you:
How large the effect is and
How uncertain the estimate is

#### Ques10. Write a Python program to calculate the margin of error for a given confidence level using sample data.


```python
def margin_of_error(data, confidence =0.95):
  data = np.array(data)
  n = len(data)

  std_err = stats.sem(data) #standard error = std dev / sqrt(n)
  t_crit = stats.t.ppf((1 + confidence) / 2 , df = n-1) #t-critical value

  moe = t_crit* std_err
  #std_Err >> It measures how much the sample mean is expected to vary from the population mean.
  #t_crit>> tells us how many standard errors we need to stretch in both directions to cover the desired confidence level.
  return moe


sample_data = [72, 75, 78, 71, 74, 77, 73, 76]
confidence_level = 0.95

moe_result = margin_of_error(sample_data, confidence_level)
print(f"Margin of Error at {confidence_level*100}% confidence: {moe_result:.3f}")
```

#### Ques10. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.



1 % people have a disease - prior = 0.01
* test:
 1. 99 % accurate is you have disease
   true_positive = 0.99  
 2. 95% accurate if you don't have
  disease.
  true_negative = 0.95    

Agar test positive aaye, to actual disease hone ki probability kya hai?


```python
def bayes_theorem(prior_disease, test_positive_if_disease, test_negative_if_no_disease):

  """
    Disease hone ki final probability calculate karta hai, agar test positive aaye.

    Parameters:
        prior_disease: Pehle se maan ke chalo ki kisi ko disease hone ka chance (e.g., 1%)
        test_positive_if_disease: Agar disease hai to test positive aane ka chance (sensitivity)
        test_negative_if_no_disease: Agar disease nahi hai to test negative aane ka chance (specificity)

    Returns:
        Final (posterior) chance ki person ko disease hai agar test positive aaya
    """

  # Disease nahi hone ka chance
  prior_no_disease = 1 - prior_disease

  # Test galti se positive aaye bina disease ke (false positive)
  test_positive_if_no_disease = 1- test_negative_if_no_disease

  #Total chance ki test positive aaye (true positive + false positive)
  total_positive_chance = (test_positive_if_disease * prior_disease) + (test_positive_if_no_disease * prior_no_disease)

  # Final answer using Bayes’ formula
  final_chance = (test_positive_if_disease * prior_disease) / total_positive_chance

  return final_chance

prior_disease = 0.01 # 1% logon ko disease
test_positive_if_disease = 0.99
test_negative_if_no_disease = 0.95

result = bayes_theorem(prior_disease, test_positive_if_disease, test_negative_if_no_disease)
print(F"Agar test positive aayte, to disease hone ka chance hai: {result : .2%}")
```
