#### Ques1. What is Simple Linear Regression.


Ans. Simple Linear Regression is a statistical method used to model the relationship between two variables:

#### Ques2. What are the key assumptions of Simple Linear Regression?

Ans.
1. Linearity: The relationship between X and Y is linear.

2. Independence of errors : The residuals (errors) are independent.One prediction‚Äôs error should not affect the error of another.

3. Homoscedasticity: variance (spread) of the errors (residuals).No matter what value X takes, the error (difference between actual and predicted Y) should not grow or shrink ‚Äî it should stay fairly even.

4. Normality of Errors: Residuals are normally distributed.

5. No Significant Outliers:
Outliers can distort slope and intercept.

#### Ques11 . What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?

Ans. Heteroscedasticity means that the spread (variance) of the residuals (errors) is not constant across all levels of the independent variables.In a good regression model, we expect homoscedasticity: equal variance of errors everywhere.
Effect of heteroscedasticity:
1. Coefficients (slopes) may still be correct, but the standard errors are wrong.

2.This leads to unreliable p-values, confidence intervals, and hypothesis tests.

3. It can mislead your interpretation of whether a variable is important or not.

#### Ques12. How can you improve a Multiple Linear Regression model with high multicollinearity?

Ans. Multicollinearity happens when two or more independent variables are highly correlated with each other.
Makes it hard to tell which variable actually affects the outcome.

Leads to unstable coefficients (values can flip signs or become huge).  
 Solutions:
 1. Remove one of the correlated variables.
 2. Combine them (e.g., average scores, sum, etc.)
 3. Use regularization techniques like Ridge or Lasso regression.

#### Ques18. What are the limitations of using R¬≤ as a sole measure of model performance?

Ans.R¬≤ tells us how well our model explains the variation in the target variable (Y). It‚Äôs useful, but if you use it alone to judge your model, it can mislead you. Here‚Äôs why:
1.  R¬≤ always increases when you add more variables, even if they‚Äôre not useful.This can make the model look better than it is (called overfitting).

2. High R¬≤ doesn‚Äôt mean the model is good at prediction. A model can fit the training data well but still perform poorly on new data.

3. R¬≤ doesn‚Äôt tell you if the relationship is real. Just because it fits doesn‚Äôt mean X is causing Y ‚Äî it might just be a coincidence.

R¬≤ is helpful, but not enough. Always look at other metrics too ‚Äî like RMSE, MAE, residual plots, and adjusted R¬≤ ‚Äî to truly understand how your model is performing.

#### Ques 28.  What are the limitations of polynomial regression?

Ans.
1. Overfitting>>
When you use a high-degree polynomial (like degree 6, 8, 10...), it may follow the training data too perfectly, including the noise and errors. This makes the model perform very well on training data but badly on new/unseen data.

2. Complexity>>
As the degree of the polynomial increases, the equation becomes longer and harder to interpret. Also, calculations become slower, especially if you have many input features.

3. Extrapolation Issues>>
Polynomial models behave unpredictably outside the range of your data.The curve might shoot up or down very fast when you try to predict values beyond what you've seen. If your data is between ages 10‚Äì50, and you try to predict for age 70 using a polynomial model ‚Äî it might give a completely nonsense result like -100 or +1000, depending on the curve.

 Polynomials bend too much outside your known data.

4. Sensitive to Outliers>>
If there are a few extreme or wrong values in your data, a polynomial model will try hard to fit them, making the curve weird and less accurate overall.

#### Ques29. What methods can be used to evaluate model fit when selecting the degree of a polynomial?

Ans. These methods help you choose the right degree:
1. Train-Test Split>>
You split your data into two parts:

 Training Data ‚Üí for building the model

 Testing Data ‚Üí for checking if it performs well on unseen data

 Then try different degrees (like 1, 2, 3, 4...) and see which one gives lowest error on the test set.

 Example:
Suppose:

 Degree 1 gives 15% error on test data

 Degree 3 gives 10% error

 egree 5 gives 8% error

 Degree 10 gives 30% error (overfitting)

 Pick degree 5, as it performs best on test data.
2. Cross-Validation (CV)>>
You divide the data into multiple parts (folds) ‚Äî like 5 or 10 equal pieces. Then:

 Train on some parts

 Test on the remaining

 Repeat this process for all combinations

 This helps you understand how the model performs on different subsets of data.

 Why it's better than simple train-test:
It gives more reliable results because it tests the model on different chunks of data ‚Äî not just one test set.

 Use CV to compare which degree consistently performs better across all data splits.
3. Evaluation Metrics>>
These metrics tell you how good your model is. You use them to compare different polynomial degrees.  
R¬≤ (R-squared): Tells how much of the data is explained by the model (closer to 1 = better).  
RMSE (Root Mean Squared Error): Lower is better; measures error in same units as output.  
MAE (Mean Absolute Error): Also lower is better; measures average absolute error
4. Learning Curves>>
ou plot model error vs. amount of training data, for both:

 Training error

 Validation error

  This helps you visually detect if the model is:

 Underfitting: Both errors are high and close

 Overfitting: Training error is low, validation error is high

 Good fit: Both errors are low and close

Try different polynomial degrees and draw learning curves. The one that shows lowest and balanced errors is the best choice

#### Ques31. How is Polynomial Regression implemented in Python?

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample data (X: independent variable, y: target variable)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 5, 10, 17, 26])  # y = x^2 + 1 approx.

# Step 1: Create polynomial features (degree 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Step 2: Fit Linear Regression on polynomial features
model = LinearRegression()
model.fit(X_poly, y)

# Step 3: Predict and visualize
y_pred = model.predict(X_poly)

# Plot
plt.scatter(X, y, color='blue', label='Original Data')
plt.plot(X, y_pred, color='red', label='Polynomial Fit')
plt.title('Polynomial Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
